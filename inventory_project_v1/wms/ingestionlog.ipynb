## Ingestion log setup

```python
%sql
create table inventory_project.metadata.ingestion_log(
    file_path STRING,
    source STRING,
    ingested_at TIMESTAMP,
    file_size BIGINT,
    status STRING
)
```

```python
from pyspark.sql.functions import current_timestamp
```

```python
row_path = '/Volumes/inventory_project/bronze/wms_raw'
source = 'wms'
files = dbutils.fs.ls(row_path)

rows = []
for f in files:
        rows.append((
        f.path,
        source,
        f.size,
        "landed"
    ))
log_df = spark.createDataFrame(rows, ["path", "source", "size", "status"])
log_df = log_df.withColumn("ingesttime", current_timestamp())

log_df.write.format("delta").mode("append").saveAsTable("inventory_project.metadata.ingestion_log")
```
